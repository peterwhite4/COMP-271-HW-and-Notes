```python
class SimpleHashTable:

    # constants for no magic values
    _DEFAULT_SIZE: int = 4
    _DEFAULT_THRESHOLD: float = 0.75
    _DEFAULT_RESIZE_BY: int = 2

    def __init__(self, size: int = _DEFAULT_SIZE) -> None:
        self._size: int = size
        self._underlying:list[None | Node] = [None] * self._size
        # how many element of the underlying array are not None
        self._occupied:int = 0
        # How many nodes are in the linked list of the underlying array
        self._count: int = 0

    def __str__(self) -> str:
        string = [f"There are {self._size} element(s) with {self._count} nodes"]
        for i in range(self._size):
            element = f"[{i:3}]: "
            head = self._underlying[i]
            while head is not None:
                element += f" --> {head.name}"
                head = head.get_next
            string.append(element)
        return string

    def add(self, name: str) -> None:
        """Assume that name is not empty and it is a legit persons name."""
        h:int = hash(name)
        # Where in the underltinh array the node with this nae will go to
        position: int = h % self._size
        # Create a node with the name
        new_node = Node(name)
        if self._underlying[position] is None:
            self._occupied += 1
        else:
            new_node.set_next(self._underlying[position])
        self._underlying[position] = new_node
        self._count += 1

    def _load_factor(self) -> float:
        return self._occupied / self._size
    

```
