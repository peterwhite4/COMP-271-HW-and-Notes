Factorials

n! = n * (n-1)!

0! = 1 BY DEFINITION

-----------
0!
1! = 1 * 0!
2! = 2 * 1!
3! = 3 * 2!
4! = 4 * 3!
5! = 5 * 4!
-----------

```python
def fact(n):
    print(f"computing {n}!", end="")
    result = 1
    if n > 0:
        print(f" as {n}x{(n-1)}!")
        result = n * fact(n-1)
    return result

# Problem with this code is that we will run out of factorials
# to look to to get the factorials, computers have limits
print()
print(fact(5))
```

Stack and Queues

QUEUE ADT
---------
+enqueue(Node): None
+dequeue: Node|None
+peek: Node|None
+is_empty: bool
+size: int
---------

dequeue - removes the head and makes the next node the head
enqueue - adds node after the tail


STACK ADT
----------
+push(Node): None
+pop: Node|None
+peek: Node|None
+is_empty: bool
+size: int
-----------

pop - removes the head; makes its next: new head
push - push data as the new head
peek - return self.__head


Both stack and queue ADTs work like Linkedlist we've seen before
